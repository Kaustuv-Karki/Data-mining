```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0, 10, 1000)
```

```python
amp = [10,10,10 ]
freq = [10,40,20]
phase = [0, np.pi, -np.pi/2]
```

```python
y = np.zeros_like(time)
```

```python
y.shape
```

```python
for amp, freq, phase in zip(amp, freq, phase):
  y += amp * np.sin(2 * np.pi * freq * time + phase ) + np.random.normal(0, 10, 1000 )
```

```python
plt.rcParams['figure.figsize'] = [20, 4]
```

```python
plt.plot(time, y)
plt.title("Sine wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
sampling_rate = 100  # Number of samples per second

fft = np.fft.fft(y)
frequencies = np.fft.fftfreq(len(fft), 1/sampling_rate)
plt.plot(frequencies, np.abs(fft))
```

```python
import pywt
```

```python
wavLt = "morl"
x = np.linspace(0,1,1000)
coeffs, freqs_cwt = pywt.cwt(y, np.arange(1, 80), wavLt, sampling_period=1/len(x))
plt.imshow(np.abs(coeffs), extent=[0, 10, freqs_cwt[-1], freqs_cwt[0]], aspect='auto', cmap='jet')
plt.colorbar(label='Magnitude')
plt.title('Wavelet Transform')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.tight_layout()#no overlappings
plt.show()
```
