```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("voice-1.csv")
```

```python
data = pd.get_dummies(data, dtype=int)
```

```python
data = data.drop(columns=["label_male"])
data.rename(columns = {'label_female':'label'}, inplace = True)
```

```python
df_y = data.iloc[:, -1]
```

```python
df_x = data.drop(columns=["label"])
```

```python
df_x.head()
```

```python
df_x.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
def standardize(x):
    x_std = StandardScaler().fit_transform(x)
    return x_std, x_std.mean(), x_std.std()
```

```python
df_x_standard, mean, std = standardize(df_x)
```

```python
df_x_standard
```

```python
df = pd.DataFrame(df_x_standard, columns = ['meanfreq', 'sd', 'median', 'Q25', 'Q75', 'IQR', 'skew', 'kurt',
       'sp.ent', 'sfm', 'mode', 'centroid', 'meanfun', 'minfun', 'maxfun',
       'meandom', 'mindom', 'maxdom', 'dfrange', 'modindx'])
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, df_y, test_size=0.2, random_state=43)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
report = classification_report(y_test, knn.predict(x_test),digits=4)
```

```python
print(report)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, knn.predict(x_test))
```

```python
plt.figure(figsize=(9, 6))
plt.title('Confusion Matrix')
sns.heatmap(cm, annot=True, fmt="", cmap='Blues',annot_kws={"fontsize": 12})
plt.xlabel('True Class')
plt.ylabel('Predicted Class')
plt.title("Confusion Matrix")
plt.figtext(0.9,0.85,'[17, 22]',ha='center',fontsize=14)
```

```python
columns
```

meanfreq: The mean frequency of the sound.
sd: The standard deviation of the sound frequency.
median: The median frequency of the sound.
Q25: The first quartile (25th percentile) frequency of the sound.
Q75: The third quartile (75th percentile) frequency of the sound.
IQR: Interquartile range, which is the difference between Q75 and Q25.
skew: Skewness, a measure of the asymmetry of the sound frequency distribution.
kurt: Kurtosis, a measure of the peakness of the sound frequency distribution.
sp.ent: Spectral entropy, which characterizes the amount of information in the sound signal.
sfm: Spectral flatness measure, which measures the flatness of the sound spectrum.
mode: The mode frequency of the sound.
centroid: The spectral centroid, a measure of the center of mass of the sound spectrum.
meanfun: The mean fundamental frequency of the sound.
minfun: The minimum fundamental frequency of the sound.
maxfun: The maximum fundamental frequency of the sound.
meandom: The mean dominant frequency of the sound.
mindom: The minimum dominant frequency of the sound.
maxdom: The maximum dominant frequency of the sound.
dfrange: The difference between maxdom and mindom.
modindx: Modulation index, which measures the degree of modulation in the sound.
label: The label or class of the sound (e.g., male or female).


```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
parameters = dict(n_neighbors = range(1,10))
```

```python
knn_clf = KNeighborsClassifier()
gscv = GridSearchCV(knn_clf, param_grid=parameters, scoring="accuracy", cv=5)
gscv.fit(x_train, y_train)
```

```python
print(gscv.best_params_)
```

```python
score = []
k_values = []
for i in range(1, 10, 2):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    k_values.append(i)
    accuracy = accuracy_score(y_test, knn.predict(x_test))
    score.append(accuracy)
```
